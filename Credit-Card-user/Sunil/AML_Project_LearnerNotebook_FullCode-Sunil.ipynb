<span class="calibre2">**Credit Card Users Churn Prediction**</span>

 

<span class="calibre5">**Problem Statement**</span>

 

<span class="calibre6">**Business Context**</span>

 

<span class="calibre7">The Thera bank recently saw a steep decline in
the number of users of their credit card,</span>

<span class="calibre7">credit cards are a good source of income for
banks because of different kinds of fees</span>

<span class="calibre7">charged by the banks like annual fees, balance
transfer fees, and cash advance fees,</span>

<span class="calibre7">late payment fees, foreign transaction fees, and
others. Some fees are charged to every</span>

<span class="calibre7">user irrespective of usage, while others are
charged under specified circumstances.</span>

 

<span class="calibre7">Customers’ leaving credit cards services would
lead bank to loss, so the bank wants to</span>

<span class="calibre7">analyze the data of customers and identify the
customers who will leave their credit</span>

<span class="calibre7">card services and reason for same – so that bank
could improve upon those areas</span>

 

<span class="calibre7">You as a Data scientist at Thera bank need to
come up with a classification model that</span>

<span class="calibre7">will help the bank improve its services so that
customers do not renounce their credit</span>

<span class="calibre7">cards</span>

 

<span class="calibre6">**Data Description**</span>

 

<span class="calibre7">CLIENTNUM: Client number. Unique identifier for
the customer holding the</span>

<span class="calibre7">account</span>

<span class="calibre7">Attrition_Flag: Internal event (customer
activity) variable - if the account is closed</span>

<span class="calibre7">then “Attrited Customer” else “Existing
Customer”</span>

<span class="calibre7">Customer_Age: Age in Years</span>

<span class="calibre7">Gender: Gender of the account holder</span>

<span class="calibre7">Dependent_count: Number of dependents</span>

<span class="calibre7">Education_Level: Educational Qualification of the
account holder - Graduate, High</span>

<span class="calibre7">School, Unknown, Uneducated, College(refers to
college student), Post-Graduate,</span>

<span class="calibre7">Doctorate</span>

<span class="calibre7">Marital_Status: Marital Status of the account
holder</span>

<span class="calibre7">Income_Category: Annual Income Category of the
account holder</span>

<span class="calibre7">Card_Category: Type of Card</span>

<span class="calibre7">Months_on_book: Period of relationship with the
bank (in months)</span>

<span class="calibre7">Total_Relationship_Count: Total no. of products
held by the customer</span>

<span class="calibre7">Months_Inactive_12_mon: No. of months inactive in
the last 12 months</span>

<span class="calibre7">Contacts_Count_12_mon: No. of Contacts in the
last 12 months</span>

<span class="calibre7">Credit_Limit: Credit Limit on the Credit
Card</span>

<span class="calibre7">Total_Revolving_Bal: Total Revolving Balance on
the Credit Card</span>

<span class="calibre7">Avg_Open_To_Buy: Open to Buy Credit Line (Average

<span class="calibre7">Total_Amt_Chng_Q4_Q1: Change in Transaction
Amount (Q4 over Q1)</span>

<span class="calibre7">Total_Trans_Amt: Total Transaction Amount (Last
12 months)</span>

<span class="calibre7">Total_Trans_Ct: Total Transaction Count (Last 12
months)</span>

<span class="calibre7">Total_Ct_Chng_Q4_Q1: Change in Transaction Count
(Q4 over Q1)</span>

<span class="calibre7">Avg_Utilization_Ratio: Average Card Utilization
Ratio</span>

 

<span class="calibre9">**What Is a Revolving Balance?**</span>

 

<span class="calibre7">If we don’t pay the balance of the revolving
credit account in full every month, the</span>

<span class="calibre7">unpaid portion carries over to the next month.
That’s called a revolving balance</span>

 

<span class="calibre7">**What is the Average Open to buy?**</span>

 

<span class="calibre7">‘Open to Buy’ means the amount left on your
credit card to use. Now, this column</span>

<span class="calibre7">represents the average of this value for the last
12 months.</span>

 

<span class="calibre7">**What is the Average utilization Ratio?**</span>

 

<span class="calibre7">The Avg_Utilization_Ratio represents how much of
the available credit the</span>

<span class="calibre7">customer spent. This is useful for calculating
credit scores.</span>

 

<span class="calibre7">**Relation b/w Avg_Open_To_Buy, Credit_Limit and
Avg_Utilization_Ratio:**</span>

 

<span class="calibre7">( Avg_Open_To_Buy / Credit_Limit ) +
Avg_Utilization_Ratio = 1</span>

<span class="calibre6">**Model evaluation criterion**</span>

 

<span class="calibre7">The nature of predictions made by the
classification model will translate as follows:</span>

 

<span class="calibre7">True Positives (TP) are existing customers
correctly predicted by the model.</span>

<span class="calibre7">True Negatives (TN) are atritioned customers
correctly predicted by the model.</span>

<span class="calibre7">False Positives (FP) are atritioned customers
incorrectly predicted as an existing</span>

<span class="calibre7">customer by the model.</span>

<span class="calibre7">False Negatives (FN) are existing customers
incorrectly predicted as an atritioned</span>

<span class="calibre7">customer by the model.</span>

 

<span class="calibre7">**Which metric to optimize?**</span>

 

<span class="calibre7">We need to choose the metric which will ensure
that the maximum number of</span>

<span class="calibre7">customer attritions are predicted correctly by
the model.</span>

<span class="calibre7">We would want Precision to be maximized as
greater the Precision, the higher the</span>

<span class="calibre7">chances of minimizing False Positives.</span>

<span class="calibre7">We want to minimize False Positives because if a
model predicts that a customer</span>

<span class="calibre7">will not attrit, but they do, the customer is
lost.</span>
